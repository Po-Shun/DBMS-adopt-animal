```python
from selenium import webdriver;
from selenium.webdriver.common.by import By;
from selenium.webdriver.support.ui import WebDriverWait;
from selenium.webdriver.support import expected_conditions as EC;
import time;

url = "https://www.pet.gov.tw/Web/O301.aspx";
```

```python
driver = webdriver.Safari();
driver.get(url);

search_button = driver.find_element_by_id("aSearch");

search_button.click();
time.sleep(2);
#WebDriverWait(driver, 50).until(EC.presence_of_element_located((By.ID, "tabPagination")));
WebDriverWait(driver, 50).until(EC.presence_of_element_located((By.ID, "USER")));

columns = [];

thead = driver.find_element_by_id("USER").find_element_by_tag_name("thead");
ths = thead.find_elements_by_tag_name("th");

for th in ths:
    columns.append(th.text);

table = [];

while (True):
    time.sleep(1);
    WebDriverWait(driver, 50).until(EC.presence_of_element_located((By.ID, "USER")));
    WebDriverWait(driver, 50).until(EC.presence_of_element_located((By.ID, "tabPagination")));
    
    next_button = driver.find_element_by_xpath("//a[@title='下一頁']");

    tbody = driver.find_element_by_id("USER").find_elements_by_tag_name("tbody")[0];
    
    trs = tbody.find_elements_by_tag_name("tr");

    for tr in trs:
        lst= [];
        tds = tr.find_elements_by_tag_name("td");
    
        for td in tds:
            lst.append(td.text);
        
        table.append(lst);
    
    next_button.click();

driver.quit();
```

```python
import pandas as pd;

df = pd.DataFrame(table, columns=columns);

df
```

```python
df.to_csv("登記機構.csv");
```
